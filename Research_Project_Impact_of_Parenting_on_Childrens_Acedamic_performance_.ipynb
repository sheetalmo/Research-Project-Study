```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder



# Step 1: Preprocess the data
# Define the target variable and features
# Here we assume 'academic_grade' as the target variable, adjust if necessary
target = 'academic_grade'
features = df.drop(columns=[target]).columns
```

```python
# Convert categorical variables to numerical
label_encoders = {}
for column in df.columns:
    if df[column].dtype == 'object':
        le = LabelEncoder()
        df[column] = le.fit_transform(df[column].astype(str))
        label_encoders[column] = le
```

```python
# Split the data into train and test sets
X = df[features]
y = df[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Step 2: Decision Tree Model
# Train the Decision Tree Classifier
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
```

```python
# Evaluate the Decision Tree
y_pred_dt = dt_model.predict(X_test)
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
print("Decision Tree Classification Report:\n", classification_report(y_test, y_pred_dt))
```

```python
# Step 3: Random Forest Model
# Train the Random Forest Classifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
```

```python
# Evaluate the Random Forest
y_pred_rf = rf_model.predict(X_test)
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Random Forest Classification Report:\n", classification_report(y_test, y_pred_rf))
```

```python
# Feature Importance
feature_importances = pd.Series(rf_model.feature_importances_, index=features).sort_values(ascending=False)
print("Feature Importances:\n", feature_importances)
```

```python
# Plot feature importances
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
feature_importances.head(10).plot(kind='barh')
plt.title("Top 10 Important Features in Random Forest")
plt.xlabel("Feature Importance Score")
plt.show()
```

# 1. Hyperparameter Tuning


```python
# increase accuracy
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for Random Forest
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

# Initialize Grid Search
grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=42),
                           param_grid=param_grid,
                           cv=5, n_jobs=-1, verbose=2, scoring='accuracy')

# Fit Grid Search
grid_search.fit(X_train, y_train)
best_rf_model = grid_search.best_estimator_

# Evaluate the tuned model
y_pred_best_rf = best_rf_model.predict(X_test)
print("Tuned Random Forest Accuracy:", accuracy_score(y_test, y_pred_best_rf))
print("Best Parameters:", grid_search.best_params_)
```

```python
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_best_rf))
print("Random Forest Classification Report:\n", classification_report(y_test, y_pred_best_rf))
```

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr, chi2_contingency, f_oneway
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Data Preprocessing
# Convert categorical variables to numerical format if needed
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace('\n', '').str.lower()  # Clean up column names
df['academic_grade'] = df['academic_grade'].astype('category').cat.codes  # Assuming 'academic_grade' is the target
```

```python
###########
# Step 2: Descriptive Statistics
# Plot to visualize general trends in academic performance and family size
sns.boxplot(data=df, x='family_members', y='academic_grade')
plt.title("Academic Performance vs Family Size")
plt.xlabel("Number of Family Members")
plt.ylabel("Academic Grade")
plt.show()
```

```python
# conclusion :

#Interpretation Example: If the correlation is negative, it might indicate that higher stress is
#associated with lower academic grades, and if significant (p < 0.05), this relationship is likely not due to random chance
```

```python
# Step 3.2: ANOVA - Effect of Number of Children on Academic Performance
# Group data by number of siblings
df['sibling'] = pd.to_numeric(df['sibling'], errors='coerce')
groups = [df[df['sibling'] == i]['academic_grade'].dropna() for i in df['sibling'].unique() if not np.isnan(i)]
anova_result = f_oneway(*groups)
print(f"ANOVA result for number of siblings affecting academic performance: F={anova_result.statistic:.2f}, p-value={anova_result.pvalue:.4f}")
```

```python
# conclusion :
# Interpretation Example: If the p-value is significant, we can conclude that the number of siblings has a meaningful impact on
#academic performance, which might suggest that larger families impact children's grades
#(positively or negatively depending on family dynamics).
```

```python
# Step 3.3: Chi-Square Test - Effect of Parental Support on Academic Performance
# Cross-tabulate between parental support and academic performance, assuming 'parent_expect_acedemic_performance' and 'academic_grade'
crosstab = pd.crosstab(df['parent_expect_acedemic_performance'], df['academic_grade'])
chi2, p, dof, expected = chi2_contingency(crosstab)
print(f"Chi-Square test for parental expectation on academic performance: Chi2={chi2:.2f}, p-value={p:.4f}")

```

```python
#Interpretation Example: A significant result might indicate that high parental expectations align with higher
#academic performance, while lower expectations correlate with lower performance, though causality cannot be inferred from
#this test alone.
```

```python
# Step 3.4: ANOVA - Emotional Stress vs. Academic Satisfaction
# Assuming 'academic_performance_satisfaction' is ordinal
df['academic_performance_satisfaction'] = pd.to_numeric(df['academic_performance_satisfaction'], errors='coerce')
groups_stress = [df[df['academic_performance_satisfaction'] == i]['emotionsl_stress'].dropna() for i in df['academic_performance_satisfaction'].unique() if not np.isnan(i)]
anova_stress_result = f_oneway(*groups_stress)
print(f"ANOVA for academic satisfaction affecting emotional stress: F={anova_stress_result.statistic:.2f}, p-value={anova_stress_result.pvalue:.4f}")
```

```python
#Interpretation Example: If significant, this result suggests that lower satisfaction with academic performance may
#correlate with higher stress levels, potentially showing the emotional impact of academic expectations.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway, chi2_contingency
from sklearn.preprocessing import LabelEncoder

# Preprocess column names for easier access
df.columns = df.columns.str.strip().str.replace(' ', '_').str.lower()
```

```python
# Visualize Home Environment Factors
# 1. Academic Performance vs. Noise Distraction at Home
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='home_distracted_by_noise', y='academic_grade')
plt.title("Academic Performance vs. Noise Distraction at Home")
plt.xlabel("Home Noise Distraction Level")
plt.ylabel("Academic Grade")
plt.show()
```

```python
# 2. Emotional Stress vs. Secure Family Environment
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='secure_family_environment', y='emotionsl_stress')
plt.title("Emotional Stress vs. Secure Family Environment")
plt.xlabel("Secure Family Environment")
plt.ylabel("Emotional Stress Level")
plt.show()
```

```python
# 3. Satisfaction vs. Family Environment Type (e.g., Supportive, Strict)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='parenting_act', y='academic_performance_satisfaction')
plt.title("Academic Satisfaction vs. Family Environment Type")
plt.xlabel("Parenting Act")
plt.ylabel("Academic Performance Satisfaction")
plt.show()
```

```python
# Statistical Tests

# Step 1: Convert categorical variables to numerical if necessary
le = LabelEncoder()
for col in ['home_distracted_by_noise', 'secure_family_environment', 'parenting_act']:
    df[col] = le.fit_transform(df[col].astype(str))
```

```python
# Step 2: Test Relationship between Home Environment and Academic Performance

# ANOVA for Home Noise Distraction Effect on Academic Grades
groups = [df[df['home_distracted_by_noise'] == level]['academic_grade'].dropna() for level in df['home_distracted_by_noise'].unique()]
anova_noise = f_oneway(*groups)
print(f"ANOVA - Effect of Home Noise Distraction on Academic Performance: F={anova_noise.statistic:.2f}, p-value={anova_noise.pvalue:.4f}")
```

```python
# ANOVA for Secure Family Environment Effect on Emotional Stress
groups_stress = [df[df['secure_family_environment'] == level]['emotionsl_stress'].dropna() for level in df['secure_family_environment'].unique()]
anova_stress = f_oneway(*groups_stress)
print(f"ANOVA - Effect of Secure Family Environment on Emotional Stress: F={anova_stress.statistic:.2f}, p-value={anova_stress.pvalue:.4f}")
```

```python
# Chi-Square Test for Parenting Act vs. Academic Satisfaction
# Cross-tabulation for categorical relationship
crosstab = pd.crosstab(df['parenting_act'], df['academic_performance_satisfaction'])
chi2, p, dof, expected = chi2_contingency(crosstab)
print(f"Chi-Square test for Parenting Act and Academic Satisfaction: Chi2={chi2:.2f}, p-value={p:.4f}")
```

```python
# Interpretation:
# For each test result, p-values < 0.05 indicate a significant effect/association,
# showing that aspects of the home environment significantly impact academic performance, emotional stress, or satisfaction.
```

```python
# Import necessary libraries
from textblob import TextBlob
import numpy as np

# Select columns for sentiment analysis
text_columns = ['stressed_level', 'emotional challenges', 'level of emotional support', "parents' emotional support reaction"]

# Preprocess and analyze sentiment
sentiment_results = {}

for col in text_columns:
    # Drop nulls, lowercase and apply TextBlob
    data[col] = data[col].astype(str).str.lower()
    data[f'{col}_sentiment'] = data[col].apply(lambda x: TextBlob(x).sentiment.polarity)

    # Categorize sentiment scores into positive, neutral, negative
    data[f'{col}_sentiment_label'] = data[f'{col}_sentiment'].apply(
        lambda x: 'positive' if x > 0 else ('negative' if x < 0 else 'neutral')
    )

    # Save sentiment distribution for each column
    sentiment_results[col] = data[f'{col}_sentiment_label'].value_counts()

# Display sentiment analysis results for each column
sentiment_results
```
